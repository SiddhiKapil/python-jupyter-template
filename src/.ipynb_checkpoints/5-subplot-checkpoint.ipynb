### Exercise 5: Subplot Layout

```python
# TASK: Create a 2x2 subplot layout.
# Plot a line chart in the first subplot, a bar chart in the second, a scatter plot in the third, and a pie chart in the fourth.
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Create figure and subplots
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# Line Chart (Top Left)
x = np.arange(0, 11, 1)
y = x ** 2
axs[0, 0].plot(x, y, marker='o', linestyle='-')
axs[0, 0].set_title('Line Chart')
axs[0, 0].set_xlabel('X Values')
axs[0, 0].set_ylabel('Y = X²')
axs[0, 0].grid(True)

# Bar Chart (Top Right)
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]
colors = ['red', 'blue', 'green', 'orange']
axs[0, 1].bar(categories, values, color=colors)
axs[0, 1].set_title('Bar Chart')
axs[0, 1].set_xlabel('Categories')
axs[0, 1].set_ylabel('Values')

# Scatter Plot (Bottom Left)
x_scatter = [1, 2, 3, 4, 5]
y_scatter = [2, 4, 6, 8, 10]
axs[1, 0].scatter(x_scatter, y_scatter, color='blue', marker='o')
for i in range(len(x_scatter)):
    axs[1, 0].annotate(f'({x_scatter[i]}, {y_scatter[i]})', (x_scatter[i], y_scatter[i]), textcoords="offset points", xytext=(5,5), ha='center')
axs[1, 0].set_title('Scatter Plot')
axs[1, 0].set_xlabel('X Values')
axs[1, 0].set_ylabel('Y Values')
axs[1, 0].grid(True)

# Pie Chart (Bottom Right)
labels = ['Python', 'Java', 'C++', 'JavaScript']
sizes = [40, 25, 20, 15]
axs[1, 1].pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90, colors=['blue', 'green', 'red', 'orange'])
axs[1, 1].set_title('Pie Chart')

# Adjust layout for better visibility
plt.tight_layout()

# Show the plots
plt.show()
```
